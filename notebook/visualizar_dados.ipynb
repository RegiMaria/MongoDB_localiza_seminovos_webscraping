```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datas = ['26/02/2024', '27/02/2024', '28/02/2024', '29/02/2024', '01/03/2024', '02/03/2024', '03/03/2024', '04/03/2024', '05/03/2024']
vendas_fiat = [66, 70, 75, 80, 85, 90, 95, 100, 105]
vendas_volkswagen = [35, 40, 45, 50, 55, 60, 65, 70, 75]
vendas_hyundai = [30, 35, 40, 45, 50, 55, 60, 65, 70]

# Configurar os dados para o gráfico:
marcas = ['FIAT', 'VOLKSWAGEN', 'HYUNDAI']
vendas_por_marca = [vendas_fiat, vendas_volkswagen, vendas_hyundai]

# Criar o gráfico de barras:
plt.figure(figsize=(10, 6))
for i, marca in enumerate(marcas):
    plt.bar(datas, vendas_por_marca[i], label=marca)

# Adicionar título e rótulos aos eixos:
plt.title('Vendas de Carros por Marca ao Longo dos Dias')
plt.xlabel('Dias')
plt.ylabel('Quantidade de Vendas')

# Adicionar legenda:
plt.legend()

# Rotacionar os rótulos do eixo x para facilitar a leitura:
plt.xticks(rotation=45)

# exibe:
plt.tight_layout()
plt.show()
```

```python
df_day1 = pd.read_csv('day1_171.csv')
df_day2 = pd.read_csv('day2_134.csv')
df_day3 = pd.read_csv('day3_128.csv')
df_day4 = pd.read_csv('day4_133.csv')
df_day5 = pd.read_csv('day5_141.csv')
df_day6 = pd.read_csv('day6_142.csv')
df_day7 = pd.read_csv('day7_139.csv')
df_day8 = pd.read_csv('day8_140.csv')
df_day9 = pd.read_csv('day9_144.csv')
df_day10 = pd.read_csv('day10_147.csv')
df_day11 = pd.read_csv('day11_150.csv')
df_day12 = pd.read_csv('day12_165.csv')
df_day13 = pd.read_csv('day13_166.csv')
df_day14 = pd.read_csv('day14_162.csv')
df_day15 = pd.read_csv('day15_171.csv')
df_day16 = pd.read_csv('day16_168.csv')
df_day17 = pd.read_csv('day17_177.csv')
df_day18 = pd.read_csv('day18_171.csv')
df_day19 = pd.read_csv('day19_167.csv')
df_day20 = pd.read_csv('day20_144.csv')


# Df e datas:
dataframes = [df_day1, df_day2, df_day3, df_day4, df_day5, df_day6, df_day7, df_day8, df_day9,df_day10, df_day11,df_day12,df_day13,df_day14,df_day15,df_day16,df_day17,df_day18,df_day19,df_day20 ]
datas = ['26/02/2024', '27/02/2024', '28/02/2024', '29/02/2024', '01/03/2024', '02/03/2024', '03/03/2024', '04/03/2024', '05/03/2024', '06/03/2024', '07/03/2024','08/03/2024', '09/03/2024','10/03/2024', '11/03/2024', '12/03/2024', '13/03/2024', '14/03/2024', '15/03/2024','16/03/2024']

# Marcas dos carros:
marcas = ['FIAT', 'VOLKSWAGEN', 'HYUNDAI', 'RENAULT', 'CHEVROLET', 'CITROEN', 'JEEP', 'TOYOTA', 'CHERY', 'PEGEOUT']
# Gráfico para cada marca:
for marca in marcas:
    # Lista para armazenar a quantidade de vendas de cada marca POR DIA:
    vendas_por_dia = []
    # Preencher a lista com a quantidade de vendas da marca PARA CADA DIA
    for df in dataframes:
        vendas_por_dia.append(df[df['marca'] == marca].shape[0])
    
    # Criar o gráfico de linha mostrando as vendas da marca ao longo dos dias:
    plt.plot(datas, vendas_por_dia, label=marca)

# Adicionar rótulos e título ao gráfico
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.title('Vendas de Carros por Marca ao Longo do Tempo')
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
